사용자로부터 `달러`, `엔`, `유로`, 또는 `위안` 금액을 입력받은 후 이를 `원`으로 변환하는 프로그램을 작성해 주세요.

각 통화별 환율은 다음과 같다.

- 달러: 1237.0
- 엔: 105.4
- 유로: 1349.69
- 위안: 194.57

사용자는 10 달러, 200 엔, 30 유로, 100 위안과 같이 금액과 통화명 사이에 **공백**을 넣어 입력한다고 가정합니다.

입력 예시) `10 달러`

결과 예시) `11670.0원`

````python
# 코드를 입력해 주세요
ex_rate={'달러': 1237.0,'엔': 105.4,'유로': 1349.69, '위안': 194.57}
amount,currency=input().split()
amount=float(amount)
rate=ex_rate.get(currency)
if rate is not None:
  money=amount*rate
  print(f"{money:.1f}원")
````

### 연습문제 2

def calc_many(choice, *args) 함수에서
- choice가 `sum` 이면 **합**을 구해서 return 합니다.
- choice 가 `mul` 이면 **곱**한 결과를 return 합니다.

함수를 구현하세요

````python
# 코드를 입력해 주세요
def calc_many(choice, *args):
  if choice=='sum':
    return sum(args)
  elif choice=='mul':
    i=1
    for nums in args:
      i*=nums
    return i
````

````python
# 결과: 10
print(calc_many('sum', 0, 1, 2, 3, 4))
# 결과: 120
print(calc_many('mul', 1, 2, 3, 4, 5))
````

## **10. 파이썬 내장함수**

### 연습문제

`zip()`과 반복문을 활용하여 다음과 같이 출력해 주세요

````python
fruit = ['사과', '자두', '딸기', '망고']
price = [1000, 3000, 2500, 6000]
````

````python
# 코드를 입력해 주세요
for f,p in zip(fruit,price):
  print("(",f,"/",p,"원)")
````
