### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
print(0 * np.nan) # NaN과의 연산은 항상 NaN을 반환한다. 0을 곱하더라도 NaN을 없앨 수 없다.
print(np.nan == np.nan) # IEEE 부동소수점 표준에 따르면, NaN은 자신을 포함해 어떤 값과도 같지 않다.
print(np.inf > np.nan) # NaN과의 비교는 항상 False를 반환한다. 비교하는 값이 무엇이든 상관없다.
print(np.nan - np.nan) # NaN과의 산술 연산은 항상 NaN을 반환한다. 따라서 NaN - NaN도 NaN이 된다.
print(np.nan in set([np.nan])) # NaN != NaN이지만, Python의 집합(set)은 객체의 동일성을 기준으로 멤버십을 결정한다. 따라서 NaN 객체는 집합에 포함된다.
print(0.3 == 3 * 0.1) # 부동소수점 연산의 정밀도 오류로 인해 0.3은 3 * 0.1과 정확히 같지 않다. 3 * 0.1의 결과는 0.3과 약간 다르다.
````


### 18. What are the result of the following expressions?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
np.array(0) / np.array(0)
np.array(0) // np.array(0)
np.array([np.nan]).astype(int).astype(float)
```

````python
print(np.array(0) / np.array(0)) # 0을 0으로 나누는 것은 정의되지 않은 연산으로, 부동소수점 연산에서 이는 NaN(Not a Number)으로 처리된다.
                                 # 이때 "invalid value encountered in true_divide" 같은 경고가 발생할 수 있다.

print(np.array(0) // np.array(0)) # 정수 나눗셈(//)에서 0을 0으로 나누는 것은 정의되지 않으며, 실제로는 NaN을 반환해야 하지만,
                                  # NumPy는 대신 0을 반환하고 경고("divide by zero encountered in floor_divide")를 표시할 수 있다.

print(np.array([np.nan]).astype(int).astype(float)) # NaN을 정수로 변환하면, 정의된 값이 없기 때문에 결과는 임의의 큰 음수 값이 된다.
                                                    # 이를 다시 부동소수점으로 변환할 때 그 값이 그대로 유지되어 부동소수점 형식으로 표시된다.
                                                    # -9.22337204e+18는 64비트 정수의 최소값에 해당하는 특수한 값이다.
````
